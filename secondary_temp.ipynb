```python
import numpy as np
import pandas as pd

# Constants / assumptions
T_BD = 300.0      # K, brown dwarf effective temperature
albedo = 0.0      # Bond albedo; 0 = absorbs everything

# Orbital separations in brown-dwarf radii
a_over_RBD = np.array([3, 5, 10, 20, 50, 100], dtype=float)

# Full heat redistribution equilibrium temperature
T_eq = T_BD * np.sqrt(1 / (2 * a_over_RBD)) * (1 - albedo)**0.25

eq_table = pd.DataFrame({
    "a / R_BD": a_over_RBD,
    "T_eq [K]": T_eq
})

eq_table
```
